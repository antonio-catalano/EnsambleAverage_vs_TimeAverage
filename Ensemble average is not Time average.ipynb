# Another brief script to show graphically the inverse relation between the variance of a series of returns and the compound return when we are in a multiplicative dynamic.

### We choose only 2 periods for brevity.

The more the variance, the greater the difference between the arithmetic average return (constant at 0) and the periodical compound return.<br>
That means that the more a market is volatile, the less the arithmetic mean reflects "the true story" of the "wealth discovery".
The more a market is volatile the more the difference between the ensemble average (arithmetic mean return) and time average (the compound return over a time trajectory).

```python
pd.options.display.float_format = '{:.1f}'.format
```

Let's build a list of ten tuples, each representing a bivariate return in two consecutive years.
The average of each tuple will be the same (zero), but the tuple's variance is going to increase.

```python
tuples = [np.array([-i,i]) for i in np.arange(0,1,0.1)]
```

In the first tuple we made 0 return in both periods, in the second tuple we made -10% in the 1st period and +10% in the 2nd period, in the third tuple we made -20% in the first period and +20% in the 2nd period, and so forth...

```python
tuples
```

```python
var_s = [np.var(elem) for elem in tuples]
cagr_s = [np.prod(elem + 1) ** (1/2) - 1 for elem in tuples]
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.plot(var_s, '--o', label = 'Variance',)
ax.plot(cagr_s, '--o', label = 'Compound return: $ r $ s.t. $C  (1 + r)^2 = FutureValue$');
ax.axhline( np.mean(tuples[0]), color = 'g', label = 'Same arithmetic average for each tuple returns ')
ax.set_ylabel('Compound return               Variance', labelpad = 18, fontsize = 14)
ax.set_xlabel('Tuple returns', labelpad = 18, fontsize = 14)
ax.xaxis.set(ticks = [i for i in range(len(tuples)) if i%2 == 0], ticklabels = [tuples[i] for i in range(len(tuples)) if i%2 == 0])
ax.tick_params(axis = 'both', which = 'major', labelsize = 13)
ax.legend(loc = 'best', fontsize = 14);
```
